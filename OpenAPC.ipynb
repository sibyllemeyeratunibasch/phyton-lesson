```python
import urllib.request
import pandas as pd
```

```python
#das ist ein langer string
open_apc_url = "https://raw.githubusercontent.com/OpenAPC/openapc-de/master/data/fuberlin/APC_FU_Berlin_2015.csv"
#in Kurzfom benennen
apc_file = "openapc.csv"
```

```python
#Befehle aus einer Library finden -> siehe Doku oder dr google -> bspw. stackoverflow ->https://stackoverflow.com/questions/22676/how-do-i-download-a-file-over-http-using-python
urllib.request.urlretrieve(open_apc_url, apc_file)
```

```python
#pandas macht tabellenformat -> cols/rows
apcs = pd.read_csv(apc_file)
```
